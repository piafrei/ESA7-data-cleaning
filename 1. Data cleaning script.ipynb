```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv')
```

```python
df
```

```python
#Checke in welchen Spalten und wie oft Null-Werte exisiteren
df.isnull().sum()
```

```python
#LÃ¶sche alle Zeilen, wo nur Null Werte exisitieren (kein inhaltlicher Mehrwert)
df = df.dropna(how='all')
```

```python
#checke ob komplette dublikate existieren (all values, exclude id)
df.duplicated(['full_name','first_name','last_name','email','gender','age'])
```

```python
#drop duplicates
df = df.drop_duplicates(['full_name','first_name','last_name','email','gender','age'])
df
```

```python
#Fill missing id with index
df.loc[df['id'].isnull(), 'id'] = df.loc[df['id'].isnull()].index + 1
df
```

```python
#check if emails have valid format
!pip install validate_email
from validate_email import validate_email

df['email'].dropna().apply(lambda x:validate_email(x))

```

```python
import numpy as np

#cleanup age
#replace all non numeric values with nan
df['age'] = pd.to_numeric(df['age'], errors='coerce')
#convert to int
df['age'].dropna().astype('int')
#replace all negative values with nan
df['age'] = df['age'].apply(lambda x : x if x > 0 else np.nan)
```
